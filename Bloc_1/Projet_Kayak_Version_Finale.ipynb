```python
import requests
import urllib.parse

df2 = pd.DataFrame()

for city in (df['City']):
    # Make a request
    r = requests.get("https://nominatim.openstreetmap.org/search?q=" + city + "&limit=1&format=json")
    # Store it in our DataFrame
    df2 = df2.append(r.json(), ignore_index=True)
```

## 2 Récuperer les données via OpenWeather avec les règles de beau temps choisies

```python
api_weather_key = '' #use your api_key
```

```python
df2['mean_temp']=''
df2['mean_prob_rain']= ''
```

```python
for i in range(len(df2)):

    lat = df2.iloc[i]['lat']
    long = df2.iloc[i]['lon']

    url = f'https://api.openweathermap.org/data/3.0/onecall?lat={lat}&lon={long}&exclude=hourly,current,minutely,alerts&units=metric&appid={api_weather_key}'
    r = requests.get(url).json()
    df_daily = pd.DataFrame(r['daily'])

    for j in range(len(r['daily'])):

      df_feelslike = pd.DataFrame(r['daily'][j]['feels_like'],index=[0])

    mean_temp = round(df_feelslike["day"].mean(),2)
    mean_prob_rain = round(df_daily["pop"].mean(),2)
    df2['mean_temp'][i]= mean_temp
    df2['mean_prob_rain'][i] = mean_prob_rain

```

```python
df2.head()
```

```python
df2['Rank'] = df2.sort_values('mean_prob_rain', ascending=True) \
               .reset_index() \
               .sort_values('index') \
               .index + 1
```

```python
df2.head()
```

```python
df2 = df2.sort_values('Rank', ascending=True)
df2.head()
```

```python
df2.to_csv('df_weather.csv')
df_best_cities = df2.head(5)
df_best_cities
```
