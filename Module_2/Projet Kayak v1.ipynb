```python
# Define a dictionary containing  data 
cities = {'City':["Mont Saint Michel",
"St Malo",
"Bayeux",
"Le Havre",
"Rouen",
"Paris",
"Amiens",
"Lille",
"Strasbourg",
"Chateau du Haut Koenigsbourg",
"Colmar",
"Eguisheim",
"Besancon",
"Dijon",
"Annecy",
"Grenoble",
"Lyon",
"Gorges du Verdon",
"Bormes les Mimosas",
"Cassis",
"Marseille",
"Aix en Provence",
"Avignon",
"Uzes",
"Nimes",
"Aigues Mortes",
"Saintes Maries de la mer",
"Collioure",
"Carcassonne",
"Ariege",
"Toulouse",
"Montauban",
"Biarritz",
"Bayonne",
"La Rochelle"]}
```

```python
import pandas as pd

# Convert the dictionary into DataFrame 
df = pd.DataFrame(cities) 
    
# Observe the result 
df 
```

```python
import requests
import urllib.parse

df2 = pd.DataFrame()

for city in (df['City']):

  rep = requests.get("https://nominatim.openstreetmap.org/search?q=" + city + "&limit=1&format=json")
  df2 = df2.append(rep.json(), ignore_index=True)
  df2 = df2[["place_id","lat", "lon"]]
  df2.insert(0, 'City', df['City'])
```

## 2 Récuperer les données via OpenWeather avec les règles de beau temps choisies

```python
df3 = pd.DataFrame()

for lati, longi in zip(df2['lat'], df2['lon']):

  r = requests.get("https://api.openweathermap.org/data/2.5/weather?lat=" + lati + "&lon=" + longi + "&exclude=current,minutely,hourly,alerts&appid=345eea7a2e2fba046096f8104b5e3506&units=metric")
  df3 = df3.append(r.json(), ignore_index=True)
```
